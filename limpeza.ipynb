## 3- Filtrando linhas necessárias 

```python
df_ceara = df_colunas_filtradas.loc[df_colunas_filtradas['SG_UF_ESC'] == 'CE'].copy()
df_ceara
```

## 4- Mapeando colunas

```python
df_ceara['TP_DEPENDENCIA_ADM_ESC'] = df_ceara['TP_DEPENDENCIA_ADM_ESC'].astype('int64')
df_ceara['TP_DEPENDENCIA_ADM_ESC'] = df_ceara['TP_DEPENDENCIA_ADM_ESC'].map({1: 'Federal', 2: 'Estadual', 3: 'Municipal', 4: 'Privada'})

df_ceara['TP_DEPENDENCIA_ADM_ESC'].unique()
```

```python
df_ceara['TP_LOCALIZACAO_ESC'] = df_ceara['TP_LOCALIZACAO_ESC'].astype('int64')
df_ceara['TP_LOCALIZACAO_ESC'] = df_ceara['TP_LOCALIZACAO_ESC'].map({1: 'Urbana', 2: 'Rural'})

df_ceara['TP_LOCALIZACAO_ESC'].unique()
```

```python
colunas = ['TP_PRESENCA_CN','TP_PRESENCA_CH', 'TP_PRESENCA_LC','TP_PRESENCA_MT']

df_ceara[colunas] = df_ceara[colunas].apply(lambda celula: celula.map({0 : 'Faltou', 1 : 'Presente', 2: 'Eliminado'} ) if celula.dtype == 'int64' else celula  )
df_ceara[colunas]
```

```python
df_filter_final = df_ceara[(df_ceara['TP_PRESENCA_CN'] == "Presente") & (df_ceara['TP_PRESENCA_CH'] == "Presente") & (df_ceara['TP_PRESENCA_LC'] == "Presente")& (df_ceara['TP_PRESENCA_MT'] == "Presente")].copy()
df_filter_final
```

```python
df_filter_final.drop(['TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC', 'TP_PRESENCA_MT'], axis=1, inplace=True)
df_filter_final
```

# Merge data

```python
from geobr import read_municipality

# Baixa dados de todos os municípios brasileiros
municipios = read_municipality(year=2020)
municipios.head()
```

```python
df_geo = df_filter_final.merge(municipios, left_on='Código do Município da Escola', right_on='code_muni', how='left')
df_geo
```

```python
df_pop = pd.read_excel('POP2024_20241230.xls', skiprows=1, sheet_name='MUNICÍPIOS', dtype={'COD. MUNIC': str})
df_pop = df_pop.loc[df_pop['UF'] == 'CE']
df_pop
```

```python
df_geo['codigo_cidade'] = df_geo['code_muni'].astype(int).astype(str).str[-5:]
df_geo['codigo_cidade'] 
```

```python
df_geo = df_geo.merge(right=df_pop[['COD. MUNIC','POPULAÇÃO ESTIMADA','NOME DO MUNICÍPIO']], how = 'left', left_on = 'codigo_cidade', right_on='COD. MUNIC')
df_geo.drop(columns = ['NOME DO MUNICÍPIO', 'COD. MUNIC'], inplace=True)
```

```python
df_geo.drop(columns=['name_state', 'code_region', 'name_region', 'codigo_cidade',
                     'code_muni', 'name_muni', 'code_state', 'abbrev_state',
                     'Número de Inscrição', 'Código da UF da Escola', 'Código do Município da Prova', 'Nome do Município da Prova',
                     'Código da UF da Prova', 'Sigla da UF da Prova'], inplace=True, errors='ignore')
df_geo
```
