```python
# Análise com IQR
def detectar_outliers_iqr(df, coluna):
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    outliers = df[(df[coluna] < limite_inferior) | (df[coluna] > limite_superior)]
    return outliers, limite_inferior, limite_superior

# Aplicar para notas
colunas_notas = ['Nota Ciências da Natureza', 'Nota Ciências Humanas', 
                 'Nota Linguagens e Códigos', 'Nota Matemática', 'Nota Redação']

for coluna in colunas_notas:
    outliers, inf, sup = detectar_outliers_iqr(df_sem_ceara, coluna)
    print(f"{coluna}:")
    print(f"  Limites: [{inf:.1f}, {sup:.1f}]")
    print(f"  Outliers encontrados: {len(outliers)}")
    print(f"  Range: [{df_sem_ceara[coluna].min()}, {df_sem_ceara[coluna].max()}]\n")
```

```python
# Análise com IQR
def detectar_outliers_iqr(df, coluna):
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    outliers = df[(df[coluna] < limite_inferior) | (df[coluna] > limite_superior)]
    return outliers, limite_inferior, limite_superior

# Aplicar para notas
colunas_notas = ['Nota Ciências da Natureza', 'Nota Ciências Humanas', 
                 'Nota Linguagens e Códigos', 'Nota Matemática', 'Nota Redação']

for coluna in colunas_notas:
    outliers, inf, sup = detectar_outliers_iqr(df_rename_ceara, coluna)
    print(f"{coluna}:")
    print(f"  Limites: [{inf:.1f}, {sup:.1f}]")
    print(f"  Outliers encontrados: {len(outliers)}")
    print(f"  Range: [{df_rename_ceara[coluna].min()}, {df_rename_ceara[coluna].max()}]\n")
```

```python
# Resumo estatístico
resumo_notas = df_rename_ceara[colunas_notas].describe()
print(resumo_notas)

# Desvio padrão para Z-score
zscore_threshold = 3  # Valores com |Z| > 3 são outliers extremos

for coluna in colunas_notas:
    from scipy import stats
    z_scores = stats.zscore(df_rename_ceara[coluna].dropna())
    outliers = z_scores[np.abs(z_scores) > zscore_threshold]
    print(f"{coluna}: {len(outliers)} outliers extremos")
```

```python
coluna = 'Nota Redação'
dados_ceara = df_rename_ceara[coluna].dropna()
dados_brasil = df_sem_ceara[coluna].dropna()

# Calcular outliers com IQR
def contar_outliers(dados):
    Q1 = dados.quantile(0.25)
    Q3 = dados.quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    outliers = dados[(dados < limite_inferior) | (dados > limite_superior)]
    return len(outliers)

outliers_ceara = contar_outliers(dados_ceara)
outliers_brasil = contar_outliers(dados_brasil)

# Criar figura com GridSpec
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(2, 2, height_ratios=[1, 0.8], hspace=0.3, wspace=0.3)

# Boxplot Ceará (esquerda)
ax1 = fig.add_subplot(gs[0, 0])
bp1 = ax1.boxplot([dados_ceara],
                   labels=['Ceará'],
                   patch_artist=True,
                   widths=0.5,
                   showmeans=True,
                   meanprops=dict(marker='D', markerfacecolor='yellow', 
                                markeredgecolor='orange', markersize=10, linewidth=2))

for patch in bp1['boxes']:
    patch.set_facecolor('#FF6B6B')
    patch.set_alpha(0.7)
    patch.set_linewidth(2)

for element in ['whiskers', 'caps', 'medians']:
    plt.setp(bp1[element], color='black', linewidth=2)

ax1.set_title('Distribuição - Ceará', fontweight='bold', fontsize=15, pad=10)
ax1.set_ylabel('Pontuação', fontweight='bold', fontsize=11)
ax1.grid(True, alpha=0.3, axis='y')
ax1.set_ylim(0, 1050)

# Boxplot Brasil (direita)
ax2 = fig.add_subplot(gs[0, 1])
bp2 = ax2.boxplot([dados_brasil],
                   labels=['Brasil'],
                   patch_artist=True,
                   widths=0.5,
                   showmeans=True,
                   meanprops=dict(marker='D', markerfacecolor='yellow', 
                                markeredgecolor='orange', markersize=10, linewidth=2))

for patch in bp2['boxes']:
    patch.set_facecolor('#4ECDC4')
    patch.set_alpha(0.7)
    patch.set_linewidth(2)

for element in ['whiskers', 'caps', 'medians']:
    plt.setp(bp2[element], color='black', linewidth=2)

ax2.set_title('Distribuição - Brasil', fontweight='bold', fontsize=13, pad=10)
ax2.set_ylabel('Pontuação', fontweight='bold', fontsize=11)
ax2.grid(True, alpha=0.3, axis='y')
ax2.set_ylim(0, 1050)

# Tabela de estatísticas (embaixo, occupando toda a largura)
ax3 = fig.add_subplot(gs[1, :])
ax3.axis('off')

# Calcular estatísticas
media_ceara = dados_ceara.mean()
media_brasil = dados_brasil.mean()
mediana_ceara = dados_ceara.median()
mediana_brasil = dados_brasil.median()
min_ceara = dados_ceara.min()
max_ceara = dados_ceara.max()
min_brasil = dados_brasil.min()
max_brasil = dados_brasil.max()
q1_ceara = dados_ceara.quantile(0.25)
q3_ceara = dados_ceara.quantile(0.75)
q1_brasil = dados_brasil.quantile(0.25)
q3_brasil = dados_brasil.quantile(0.75)

# Criar dados da tabela
table_data = [
    ['Métrica', 'Ceará', 'Brasil', 'Diferença'],
    ['Média', f'{media_ceara:.2f}', f'{media_brasil:.2f}', f'{media_brasil - media_ceara:+.2f}'],
    ['Mediana', f'{mediana_ceara:.2f}', f'{mediana_brasil:.2f}', f'{mediana_brasil - mediana_ceara:+.2f}'],
    ['Mínimo', f'{min_ceara:.2f}', f'{min_brasil:.2f}', f'{min_brasil - min_ceara:+.2f}'],
    ['Máximo', f'{max_ceara:.2f}', f'{max_brasil:.2f}', f'{max_brasil - max_ceara:+.2f}'],
    ['Q1', f'{q1_ceara:.2f}', f'{q1_brasil:.2f}', f'{q1_brasil - q1_ceara:+.2f}'],
    ['Q3', f'{q3_ceara:.2f}', f'{q3_brasil:.2f}', f'{q3_brasil - q3_ceara:+.2f}'],
    ['Amostra (N)', f'{len(dados_ceara):,}', f'{len(dados_brasil):,}', f'{len(dados_brasil) - len(dados_ceara):+,}'],
    ['Outliers', f'{outliers_ceara:,}', f'{outliers_brasil:,}', f'{outliers_brasil - outliers_ceara:+,}']
]

# Crear table
table = ax3.table(cellText=table_data, 
                  cellLoc='center',
                  loc='center',
                  colWidths=[0.2, 0.25, 0.25, 0.25],
                  bbox=[0.05, 0.1, 0.9, 0.9])

table.auto_set_font_size(False)
table.set_fontsize(15)
table.scale(1, 2.5)

# Colorir cabeçalho
for i in range(4):
    table[(0, i)].set_facecolor('#4A90E2')
    table[(0, i)].set_text_props(weight='bold', color='white')

# Colorir linhas alternadas
for i in range(1, len(table_data)):
    for j in range(4):
        if i % 2 == 0:
            table[(i, j)].set_facecolor('#F0F0F0')
        else:
            table[(i, j)].set_facecolor('white')

# Título geral
fig.suptitle(f"{coluna} - Análise Comparativa Ceará vs Brasil (ENEM 2023)", 
             fontsize=18, fontweight='bold', y=0.98)

plt.tight_layout()
plt.show()
```
